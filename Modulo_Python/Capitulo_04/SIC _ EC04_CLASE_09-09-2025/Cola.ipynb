```python
#   DECLARACIÓN DE UNA COLA EN PYTHON

cola = []   # cola vacía
print("Cola vacía:", cola)

#   OPERACIONES BÁSICAS DE UNA COLA

# ENQUEUE → agregar un elemento al final de la cola
cola.append("A")
cola.append("B")
cola.append("C")
print("Después de enqueue A, B, C:", cola)  # ['A', 'B', 'C']

# DEQUEUE → quitar y devolver el elemento del frente (posición 0)
primero = cola.pop(0)
print("Elemento atendido (dequeue):", primero)
print("Cola después de dequeue:", cola)

# IS_EMPTY → comprobar si la cola está vacía
esta_vacia = (len(cola) == 0)
print("¿La cola está vacía?", esta_vacia)


#   FUNCIONES DE UTILIDAD

def enqueue(cola, elem):
    """Agrega un elemento al final de la cola"""
    cola.append(elem)

def dequeue(cola):
    """
    Extrae y devuelve el frente de la cola.
    Lanza IndexError si la cola está vacía.
    """
    return cola.pop(0)

def is_empty(cola):
    """Devuelve True si la cola está vacía"""
    return len(cola) == 0

def size(cola):
    """Devuelve la cantidad de elementos en la cola"""
    return len(cola)


#   VISUALIZACIÓN DE LA COLA

def imprimir_cola(cola):
    """
    Muestra la cola de forma visual:
    [frente] → ... → [final]
    """
    if not cola:
        print("[cola vacía]")
    else:
        vista = " -> ".join(str(x) for x in cola)
        print("Frente →", vista, "← Final")

#   EJEMPLO: SISTEMA DE ATENCIÓN EN UN BANCO


def banco_simulacion():
    cola = []

    # Llegan tres clientes
    print("\nLlegan tres clientes: Ana, Luis y Marta")
    enqueue(cola, "Ana")
    enqueue(cola, "Luis")
    enqueue(cola, "Marta")
    imprimir_cola(cola)

    # Atienden al primero
    print("\nAtendiendo a:", dequeue(cola))
    imprimir_cola(cola)

    # Llegan dos clientes más
    print("\nLlegan dos clientes: Pedro y Sofía")
    enqueue(cola, "Pedro")
    enqueue(cola, "Sofía")
    imprimir_cola(cola)

    # Atienden al siguiente
    print("\nAtendiendo a:", dequeue(cola))
    imprimir_cola(cola)

    # Atienden al siguiente
    print("\nAtendiendo a:", dequeue(cola))
    imprimir_cola(cola)

    # Atienden al siguiente
    print("\nAtendiendo a:", dequeue(cola))
    imprimir_cola(cola)

    # Atienden al último
    print("\nAtendiendo a:", dequeue(cola))
    imprimir_cola(cola)

    print("\nTodos los clientes han sido atendidos ")


if __name__ == "__main__":
    banco_simulacion()
```
