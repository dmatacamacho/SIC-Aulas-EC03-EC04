```python

#   TEMA: BÚSQUEDA SECUENCIAL 


#   FUNCIÓN DE BÚSQUEDA SECUENCIAL

def busqueda_secuencial(lista, x, mostrar_pasos=False):
    """
    Busca el valor x en la lista.
    Devuelve el índice si lo encuentra, -1 si no está.
    Si mostrar_pasos=True, imprime cada comparación.
    """
    for i, valor in enumerate(lista):  # recorremos la lista con índice y valor
        if mostrar_pasos:
            print(f"Comparando x={x} con lista[{i}]={valor}")
        if valor == x:
            return i   # encontrado
    return -1          # no encontrado


#   OTRA APLICACIÓN: BUSCAR EL MAYOR ELEMENTO

def buscar_mayor(lista, mostrar_pasos=False):
    """
    Encuentra el valor más grande en la lista y devuelve su índice.
    """
    if not lista:
        return -1  # lista vacía
    indice_mayor = 0  # asumimos que el mayor está en la posición 0
    for i in range(1, len(lista)):
        if mostrar_pasos:
            print(f"Comparando lista[{i}]={lista[i]} con lista[{indice_mayor}]={lista[indice_mayor]}")
        if lista[i] > lista[indice_mayor]:
            indice_mayor = i
    return indice_mayor

#   VISUALIZACIÓN DE LA LISTA

def imprimir_lista(lista):
    """
    Muestra la lista de forma clara con índices.
    """
    print("Índices:", " ".join(f"{i:2}" for i in range(len(lista))))
    print("Valores:", " ".join(f"{v:2}" for v in lista))


#   PRUEBAS DEL ALGORITMO

if __name__ == "__main__":
    # Lista de ejemplo
    numeros = [4, 8, 2, 7, 5]

    print("\nLista de números:")
    imprimir_lista(numeros)

    # Buscar un valor que sí está
    print("\n--- Búsqueda de 7 ---")
    pos = busqueda_secuencial(numeros, 7, mostrar_pasos=True)
    print("Resultado:", "Encontrado en índice" if pos != -1 else "No encontrado", pos)

    # Buscar un valor que no está
    print("\n--- Búsqueda de 9 ---")
    pos = busqueda_secuencial(numeros, 9, mostrar_pasos=True)
    print("Resultado:", "Encontrado en índice" if pos != -1 else "No encontrado", pos)

    # Buscar el mayor elemento
    print("\n--- Buscar el mayor elemento ---")
    indice = buscar_mayor(numeros, mostrar_pasos=True)
    print(f"El mayor es {numeros[indice]} en el índice {indice}")
```
