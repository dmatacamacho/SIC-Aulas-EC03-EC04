Su tarea será **analizar esta Serie** aplicando operaciones estadísticas y visualizaciones para entender mejor los hábitos de uso.

3. **Estadísticas básicas**

5. **Visualizaciones**

   * ¡Se creativo! Elige los gráficos que mejor expliquen los resultados.


```python
import random
import pandas as pd
import matplotlib.pyplot as plt
```

### 1. **Crear la Serie**

```python
days = [f"Día {x}" for x in range(1, 31)]
hoursQuantity = pd.Series([random.randint(0, 10) for _ in range(0, 30)],
                          index= days,
                          name='Uso de redes sociales (horas)')
```

### 2. **Exploración de la Serie**

```python
hoursQuantity.iloc[:5]
```

```python
hoursQuantity.iloc[-5:]
```

```python
print(f"Día 7: {hoursQuantity.iloc[7]} horas")
print(f"Longitud: {len(hoursQuantity)}")
print(f"Tipo de datos: {type(hoursQuantity)}")
print(f"Datos nulos: {hoursQuantity.isnull().sum()}")
```

### 3. **Estadísticas básicas**

```python
print(f"Promedio del uso de redes sociales: {hoursQuantity.mean()} horas")
print(f"La mediana del uso de redes sociales: {hoursQuantity.median()} horas")
print(f"El mínimo uso de redes sociales: {hoursQuantity.min()} horas")
print(f"El máximo uso de redes sociales: {hoursQuantity.max()} horas")
print(f"Varianza del uso de redes sociales: {hoursQuantity.var():.2f} horas")
print(f"Desviación estándar del uso de redes sociales: {hoursQuantity.std():.2f} horas")
print(f"Día con menor uso de redes sociales: {hoursQuantity.idxmin()} con {hoursQuantity[hoursQuantity.idxmin()]} horas")
print(f"Día con nayor uso de redes sociales: {hoursQuantity.idxmax()} con {hoursQuantity[hoursQuantity.idxmax()]} horas")
```

### 4. **Filtrado y condiciones**

```python
print("Días donde se usaron más de 6 horas.")
hoursQuantity.loc[hoursQuantity > 6]
```

```python
print("Conteo de días donde se usaron menos de 2 horas: ")
print(hoursQuantity.loc[hoursQuantity < 2].count())
hoursQuantity.loc[hoursQuantity < 2]
```

### 5. **Visualizaciones**

```python
plt.style.use('seaborn-v0_8')

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

fig.suptitle('Análisis de horas de uso de redes sociales', fontsize=16, fontweight='bold')

hoursQuantity.plot(kind='line', marker='o', ax=ax1, color='blue')
ax1.set_title('Evolución diaria')
ax1.set_ylabel('Horas')
ax1.tick_params(axis='x', rotation=45)

hoursQuantity.plot(kind='bar', ax=ax2, color='green', edgecolor='black')
ax2.set_title('Comparación día a día')
ax2.set_ylabel('Horas')
ax2.tick_params(axis='x', rotation=90)

ax3.scatter(range(1, 31), hoursQuantity, color='orange', alpha=0.7, edgecolor='black')
ax3.set_title('Dispersión por día')
ax3.set_xlabel('Día')
ax3.set_ylabel('Horas')
ax3.grid(True, alpha=0.3)

hoursQuantity.plot(kind='hist', bins=11, rwidth=0.8, ax=ax4, color='purple', edgecolor='black', alpha=0.7)
ax4.set_title('Distribución de horas')
ax4.set_xlabel('Horas')
ax4.set_ylabel('Frecuencia')
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
