# Ejercicio: Análisis de horas de uso de redes sociales

Un grupo de estudiantes registró durante 30 días la **cantidad de horas que pasaron usando redes sociales**.
Los datos están representados en una **Serie de Pandas**, donde cada índice corresponde al día (`Día 1`, `Día 2`, ..., `Día 30`).

Su tarea será **analizar esta Serie** aplicando operaciones estadísticas y visualizaciones para entender mejor los hábitos de uso.

## Instrucciones

1. **Crear la Serie**

   * Genera una lista de 30 valores (pueden ser aleatorios o inventados entre 0 y 10 horas).
   * Convierte la lista en una Serie de Pandas.
   * Usa como índice `Día 1`, `Día 2`, ..., `Día 30`.

2. **Exploración de la Serie**

   * Muestra los primeros y últimos 5 valores.
   * Consulta el valor de un día específico (por ejemplo, `"Día 7"`).
   * Verifica la longitud, el tipo de datos y si existen valores faltantes.

3. **Estadísticas básicas**

   * Calcula: promedio, mediana, mínimo, máximo, varianza y desviación estándar.
   * Identifica los días con mayor y menor uso de redes sociales.

4. **Filtrado y condiciones**

   * Filtra los días donde se usaron **más de 6 horas**.
   * Cuenta cuántos días el uso fue **menor a 2 horas**.


5. **Visualizaciones**

   * Realiza al menos **2 gráficos diferentes** para representar la información.

   * ¡Se creativo! Elige los gráficos que mejor expliquen los resultados.


```python
import random
import pandas as pd

dias = [f"Dia {i}" for i in range(1, 31)]
horas = [random.randint(0, 10) for i in range(30)]

serie = pd.Series(horas, index=dias, name="Número de horas por día")

print(serie)
```

```python
# Exploracion de la serie
print("Primeros 5 valores:")    # 5 primeros valores
print(serie.head(5))

print("\nÚltimos 5 valores:")    #5 ultimso valores
print(serie.tail(5))

#consultar el valor de una dia especifico
valor_dia7 = serie["Dia 7"]
print("Número de horas en Día 7:", valor_dia7)

#para que el usuario consulte cualquier dia
dia_usuario = input("Escribe el día que quieres consultar (por ejemplo: Dia 7): ")

if dia_usuario in serie.index:
    print(f"Número de horas en {dia_usuario}: {serie[dia_usuario]}")
else:
    print("Ese día no existe en la serie.")


# para longitud, tipo de datos y valores faltantes
print("Longitud de la serie:", len(serie))

print("Tipo de datos de la serie:", serie.dtype)

print("¿Existen valores faltantes?:", serie.isnull().any())
```

```python
#Estadisticas basicas

print("\n--- Estadisticas basicas ---")

#Medidas principales
promedio = serie.mean()
mediana = serie.median()
minimo = serie.min()
maximo = serie.max()
varianza = serie.var()
desviacion_estandar = serie.std()

print(f"Promedio de horas: {promedio}")
print(f"Mediana de horas: {mediana}")
print(f"Valor mínimo de horas: {minimo}")
print(f"Valor máximo de horas: {maximo}")
print(f"Varianza de horas: {varianza}")
print(f"Desviación estándar de horas: {desviacion_estandar}")

# Identificar los dìas con mayor y menor uso

dia_max = serie.idxmax()
dia_min = serie.idxmin()

print(f"\n El Día con mayor uso de redes sociales fue : {dia_max} con {maximo} horas.")
print(f"El día con menor uso de redes sociales fue {dia_min} con {minimo} horas.")
```

```python
print("\n" + "="*50)

print("Filtrado y condiciones")

print ("="*50)

# Filtra los días donde se usaron más de 6 horas.
dias_mayor_6 = serie[serie > 6]
print("\ndías donde se usaron más de 6 horas:")
print(dias_mayor_6)

# Cuenta cuántos días el uso fue menor a 2 horas.
dias_menor_2 = (serie < 2).sum()
print(f"\ndías el uso fue menor a 2 horas: {dias_menor_2}")
```

```python
import matplotlib.pyplot as plt
from matplotlib.markers import MarkerStyle

plt.style.use('dark_background')
plt.figure(figsize=(10, 5))
serie.plot(marker='$\u266B$', linestyle='dashdot', color='gold', markerfacecolor='snow',
           markersize=20, linewidth=2, markeredgecolor='none')
plt.title('horas de RS en el mes', color='mediumseagreen', fontsize=20)
plt.xlabel('Día', color='azure', fontsize=18)
plt.ylabel('Horas', color='cyan', fontsize=18)
plt.xticks(rotation=45, color='lightcoral')
plt.yticks(color='orange')
plt.grid(True, alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 4))
plt.hist(serie, bins=range(0, 12), edgecolor='cornflowerblue', linewidth=6, alpha=0.8, color='indigo')
plt.title('Horas por dia en RS', color='gold', fontsize=14)
plt.xlabel('Horas', color='turquoise', fontsize=12)
plt.ylabel('Frecuencia', color='turquoise', fontsize=12)
plt.xticks(range(0, 11), color='white')
plt.yticks(color='white')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
