# Ejercicio: Análisis de horas de uso de redes sociales

Un grupo de estudiantes registró durante 30 días la **cantidad de horas que pasaron usando redes sociales**.
Los datos están representados en una **Serie de Pandas**, donde cada índice corresponde al día (`Día 1`, `Día 2`, ..., `Día 30`).

Su tarea será **analizar esta Serie** aplicando operaciones estadísticas y visualizaciones para entender mejor los hábitos de uso.

## Instrucciones

1. **Crear la Serie**

   * Genera una lista de 30 valores (pueden ser aleatorios o inventados entre 0 y 10 horas).
   * Convierte la lista en una Serie de Pandas.
   * Usa como índice `Día 1`, `Día 2`, ..., `Día 30`.

2. **Exploración de la Serie**

   * Muestra los primeros y últimos 5 valores.
   * Consulta el valor de un día específico (por ejemplo, `"Día 7"`).
   * Verifica la longitud, el tipo de datos y si existen valores faltantes.

3. **Estadísticas básicas**

   * Calcula: promedio, mediana, mínimo, máximo, varianza y desviación estándar.
   * Identifica los días con mayor y menor uso de redes sociales.

4. **Filtrado y condiciones**

   * Filtra los días donde se usaron **más de 6 horas**.
   * Cuenta cuántos días el uso fue **menor a 2 horas**.


5. **Visualizaciones**

   * Realiza al menos **2 gráficos diferentes** para representar la información.
   * Puedes elegir entre:

   * ¡Se creativo! Elige los gráficos que mejor expliquen los resultados.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Creando Series**

```python
# Generando 30 valores aleatorios entre 0 y 10 horas
dias = ['Día ' + str(i) for i in range(1, 31)]
horas = np.random.randint(0, 11, 30)

# Convertir la lista en una Serie de Pandas.
serie = pd.Series(horas, index=dias, name="Uso redes sociales (horas)")
print(serie)


# Usar como índice Día 1, Día 2, ..., Día 30.
```

**Explorando las series**

```python
# Primeros y últimos 5 valores
print(serie.head())

print(serie.tail())

# Consultando un valor específico
print(serie["Día 7"])

# Verificando la longitud, el tipo de datos y si existen valores faltantes.
print("Longitud:", len(serie))
print("Tipo de datos:", serie.dtype)
print("Valores nulos:", serie.isnull().sum())
```

**Estadísticas Básicas**

```python
# Calculando: promedio, mediana, mínimo, máximo, varianza y desviación estándar.
# Identificando los días con mayor y menor uso de redes sociales.

print("Promedio:", serie.mean())
print("Mediana:", serie.median())
print("Mínimo:", serie.min())
print("Máximo:", serie.max())
print("Varianza:", serie.var())
print("Desviación estándar:", serie.std())

print("Día de mayor uso:", serie.idxmax(), "con", serie.max(), "horas")
print("Día de menor uso:", serie.idxmin(), "con", serie.min(), "horas")
```

**Filtrado y condiciones**

```python
print("Dias con más de 6 horas")
print(serie[serie > 6])

print("Cantidad de dias con menos de 2 horas:", (serie < 2).sum())
```

**Visualizaciones**

```python

# grafico de puntos
plt.figure(figsize=(12,6))
plt.scatter(serie.index, serie.values, c=serie.values, cmap="viridis", s=80)
plt.title("Uso diario de redes sociales")
plt.xticks(rotation=90)
plt.ylabel("Horas")
plt.plot(serie.index, serie.values, color="gray", linestyle="--", alpha=0.5)
plt.show()

# mapa de calor
valores = serie.values
faltan = (7 - (len(valores) % 7)) % 7
if faltan > 0:
    valores = np.append(valores, [np.nan]*faltan)
data = valores.reshape(-1, 7)
plt.figure(figsize=(12,6))
sns.heatmap(data, annot=True, cmap="YlGnBu", linewidths=1, linecolor="white", cbar_kws={'label': 'Horas'})
plt.title("Mapa de calor de uso de redes sociales")
plt.xlabel("Días de la semana")
plt.ylabel("Semana")
plt.xticks(np.arange(7)+0.5, ["Lun","Mar","Mié","Jue","Vie","Sáb","Dom"])
plt.show()

# grafico de barras con color dependiente de las horas
umbral = 6
binario = np.where(data > umbral, 1, 0)
plt.figure(figsize=(12,6))
sns.heatmap(binario, annot=data, fmt=".0f", cmap=["lightgreen","red"],
            cbar=False, linewidths=1, linecolor="white")
plt.title("Mapa binario")
plt.xlabel("Días de la semana")
plt.ylabel("Semana")
plt.xticks(np.arange(7)+0.5, ["Lun","Mar","Mié","Jue","Vie","Sáb","Dom"])
plt.show()

# grafico de barras
frecuencia = serie.value_counts().sort_values(ascending=False)
plt.figure(figsize=(12,6))
plt.bar(frecuencia.index.astype(str), frecuencia.values)
plt.title("Frecuencia de horas de uso")
plt.xlabel("Horas")
plt.ylabel("Frecuencia")
plt.show()
```
