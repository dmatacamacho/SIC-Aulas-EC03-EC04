#Crear serie

```python
import random
import pandas as pd

dias = [f"Dia {i}" for i in range(1, 31)]
print(dias)

random.seed(100)
horas = [random.randint(0, 10) for _ in range(30)]
print(horas)

uso_redes = pd.Series(horas, index=dias)
print(uso_redes)
```

# Explorar serie

```python
#Exploración de datos

#Muestra los primeros y últimos 5 valores
print(uso_redes.head())
print(uso_redes.tail())

#Consulta el valor de un día específico (por ejemplo, `"Día 10"`)
print(uso_redes["Dia 10"])

#Verifica la longitud, el tipo de datos y si existen valores faltantes
print(len(uso_redes))
print(uso_redes.dtype)
print(uso_redes.isnull().sum())
```

# Estadísticas básicas

```python
#Estadísticas básicas
#Calcula: promedio, mediana, mínimo, máximo, varianza y desviación estándar
print(uso_redes.mean())
print(uso_redes.median())
print(uso_redes.min())
print(uso_redes.max())
print(uso_redes.var())
print(uso_redes.std())

#Identifica los días con mayor y menor uso de redes sociales
print(uso_redes.idxmax(), uso_redes.max())
print(uso_redes.idxmin(), uso_redes.min())
```

# Filtrado y condiciones

```python
#Filtrado y condiciones
#Filtra los días donde se usaron mas de 6 horas
print(uso_redes[uso_redes > 6])
#Cuenta cuántos días el uso fue menor a 2 horas
print((uso_redes < 2).sum())
```

# Visualización

```python
import sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
import scienceplots


#Como las series son un arreglo unidimensional tenemos que convertirlas en una matriz
info_matriz = uso_redes.values.reshape(5,6) #como son 30 datos lo escribimos como la multimplicación 5x6
#creamos las etiquetas para nuestro gráfico
dias_matriz=np.array(dias).reshape(5,6)
#Graficamos el heat map
plt.figure(figsize=(8, 5))
sns.heatmap(info_matriz, annot=dias_matriz, fmt="", cmap="YlOrRd", cbar=True)
plt.title("Uso de Redes Sociales (Horas por Dia)")
plt.show()
```

```python
categorias = {
    "Bajo (<2h)": (uso_redes < 2).sum(),
    "Moderado (2-6h)": ((uso_redes >= 2) & (uso_redes <= 6)).sum(),
    "Alto (>6h)": (uso_redes > 6).sum()
}

plt.figure(figsize=(6,6))
plt.pie(categorias.values(), labels=categorias.keys(), autopct="%1.1f%%",
        startangle=90, colors=["lightgreen","gold","salmon"])
plt.title("Categorías de uso de redes sociales")
plt.show()
```

```python
# media y mediana
media = uso_redes.mean()
mediana = uso_redes.median()

# Definir colores para las barras
colors = ['salmon' if uso > 6 else ('lightgreen' if uso < 3 else 'bisque') for uso in uso_redes]

plt.figure(figsize=(12,6))
plt.bar(uso_redes.index, uso_redes.values, color=colors)

# Líneas de media y mediana
linea_media = plt.axhline(media, color="red", linestyle="--", linewidth=2, label=f"Media: {media:.2f}")
linea_mediana = plt.axhline(mediana, color="green", linestyle="-.", linewidth=2, label=f"Mediana: {mediana:.2f}")

plt.title("Horas de uso de redes sociales por día (30 días)", fontsize=16, fontweight="bold")
plt.xlabel("Días", fontsize=12)
plt.ylabel("Horas de uso", fontsize=12)
plt.xticks(rotation=45)
plt.ylim(0, 11)  # límite hasta 10 horas

plt.grid(axis='y', alpha=0.3)

# leyenda por colores
red_patch = mpatches.Patch(color='salmon', label='Uso > 6 horas')
green_patch = mpatches.Patch(color='lightgreen', label='Uso < 3 horas')
yellow_patch = mpatches.Patch(color='bisque', label='Uso entre 3 y 6 horas')

# Combinar parches y líneas en la leyenda
plt.legend(handles=[red_patch, green_patch, yellow_patch, linea_media, linea_mediana],
           labels=['Uso > 6 horas', 'Uso < 3 horas', 'Uso entre 3 y 6 horas',
                   f"Media: {media:.2f}", f"Mediana: {mediana:.2f}"],
           loc="upper left", fontsize=8)


plt.show()
```

```python
plt.figure(figsize=(12,6))

# Dibujar línea general
plt.plot(uso_redes.index, uso_redes.values, color="lightgray", linewidth=1, zorder=1)

# Colorear puntos según condición
for dia, uso in zip(uso_redes.index, uso_redes.values):
    if uso > 6:
        plt.scatter(dia, uso, color="salmon", s=80, zorder=2, label="Uso > 6 horas" if dia == uso_redes.index[0] else "")
    elif uso < 3:
        plt.scatter(dia, uso, color="lightgreen", s=80, zorder=2, label="Uso < 3 horas" if dia == uso_redes.index[0] else "")
    else:
        plt.scatter(dia, uso, color="bisque", s=80, zorder=2, label="Uso entre 3 y 6 horas" if dia == uso_redes.index[0] else "")

# Líneas de media y mediana
linea_media = plt.axhline(media, color="red", linestyle="--", linewidth=2, label=f"Media: {media:.2f}")
linea_mediana = plt.axhline(mediana, color="green", linestyle="-.", linewidth=2, label=f"Mediana: {mediana:.2f}")

plt.title("Horas de uso de redes sociales por día (30 días)", fontsize=16, fontweight="bold")
plt.xlabel("Días", fontsize=12)
plt.ylabel("Horas de uso", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

# Parches personalizados para la leyenda de colores
red_patch = mpatches.Patch(color='salmon', label='Uso > 6 horas')
green_patch = mpatches.Patch(color='lightgreen', label='Uso < 3 horas')
yellow_patch = mpatches.Patch(color='bisque', label='Uso entre 3 y 6 horas')

# Combinar parches y líneas en la leyenda
plt.legend(handles=[red_patch, green_patch, yellow_patch, linea_media, linea_mediana],
           labels=['Uso > 6 horas', 'Uso < 3 horas', 'Uso entre 3 y 6 horas',
                   f"Media: {media:.2f}", f"Mediana: {mediana:.2f}"],
           loc="upper left", fontsize=8)

plt.show()
```
