# Ejercicio: Análisis de horas de uso de redes sociales

Un grupo de estudiantes registró durante 30 días la **cantidad de horas que pasaron usando redes sociales**.
Los datos están representados en una **Serie de Pandas**, donde cada índice corresponde al día (`Día 1`, `Día 2`, ..., `Día 30`).

Su tarea será **analizar esta Serie** aplicando operaciones estadísticas y visualizaciones para entender mejor los hábitos de uso.

---

## Instrucciones

1. **Crear la Serie**

   * Genera una lista de 30 valores (pueden ser aleatorios o inventados entre 0 y 10 horas).
   * Convierte la lista en una Serie de Pandas.
   * Usa como índice `Día 1`, `Día 2`, ..., `Día 30`.

2. **Exploración de la Serie**

   * Muestra los primeros y últimos 5 valores.
   * Consulta el valor de un día específico (por ejemplo, `"Día 7"`).
   * Verifica la longitud, el tipo de datos y si existen valores faltantes.

3. **Estadísticas básicas**

   * Calcula: promedio, mediana, mínimo, máximo, varianza y desviación estándar.
   * Identifica los días con mayor y menor uso de redes sociales.

4. **Filtrado y condiciones**

   * Filtra los días donde se usaron **más de 6 horas**.
   * Cuenta cuántos días el uso fue **menor a 2 horas**.


5. **Visualizaciones**

   * Realiza al menos **2 gráficos diferentes** para representar la información.
   * Puedes elegir entre:

     * Gráfico de línea para ver la evolución diaria.
     * Histograma para la distribución de horas.
     * Gráfico de barras para comparar ciertos días.
   * ¡Se creativo! Elige los gráficos que mejor expliquen los resultados.


```python
"""1. Crear serie (con diccionarios)"""
import random as rd
import pandas as pd
clave = "Día "
uso_redes = dict()

for i in range(1,31):
  uso_redes[clave+str(i)] = rd.randint(0,10)

serie_redes = pd.Series(uso_redes, name = "Uso de redes diario (Horas)")
print(serie_redes)
```

```python
"""2. Exploración de la serie"""
import pandas as pd

def explorar(serie_redes):
  print(f"Primeros 5 valores:\n{serie_redes.iloc[0:5]}")
  print(f"Últimos 5 valores:\n{serie_redes.iloc[-5:]}")
  ind = int(input("Número del día que quiere buscar: "))
  print(f"Horas de aquel dia: {serie_redes.loc['Día '+str(ind)]}")
  print(f"Longitud:{len(serie_redes)}")
  print(f"Tipo de datos: {serie_redes.dtypes}")
  print(f"Hay valores faltantes: {'No hay ' if serie_redes.isnull().any()== False else 'Si hay' }")

explorar(serie_redes)


```

```python
"""3. Estadísticas """
#identificar los días de mayor y menor uso de redes
mayor_uso = []
menor_uso = []
for clave,valor in uso_redes.items():
  if uso_redes[clave] == int(serie_redes.min()):
    menor_uso.append(clave)
  if uso_redes[clave] == int(serie_redes.max()):
    mayor_uso.append(clave)
print("="*40)
print("="*8, "Estadísticas Mensuales", "="*8)
print("="*40)
print(f"\n•Promedio de uso diario: {serie_redes.mean():.3f} horas")

print(f"\n•Uso mediano de redes: {serie_redes.median():.2f} horas")

print(f"\n•Mínimo de horas diario: {serie_redes.min()} Horas.\n   +Esto ocurrió en {menor_uso}")

print(f"\n•Máximo de horas diario: {serie_redes.max()} Horas.\n   +Esto ocurrió en {mayor_uso}")

print(f"\n•Variación de uso: {serie_redes.var():.2f}")

print(f"\n•Desviación standard:{serie_redes.std():.2f}")

print(f"\n•Total de horas al mes: {serie_redes.sum():}\n   +Esto significa que pasaste un {((serie_redes.sum())/(31*24))*100:.2f}% de tu tiempo al mes en las redes!")
```

```python
"""4. Filtrado y condiciones"""
# Días con más de seis horas
mas_seis_horas = serie_redes[serie_redes > 6]
print(mas_seis_horas)

# Días con menos de dos horas
menos_dos_horas = serie_redes[serie_redes < 2]
print("La cantidad de días donde el uso fue de menos de dos horas son", len(menos_dos_horas))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))
plt.plot(serie_redes.index, serie_redes.values, marker="o", color="royalblue")
plt.title("Evolución del uso de redes sociales (30 días)")
plt.xlabel("Días")
plt.ylabel("Horas de uso")
plt.xticks(serie_redes.index, rotation=45)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
media = serie_redes.mean()
print(f"Media de horas de uso: {media:.2f}")
dias_mayor_consumo = serie_redes[serie_redes > media]
df = dias_mayor_consumo.reset_index()
df.columns = ["Día", "Horas"]
df["NumDia"] = df["Día"].str.replace("Día ", "").astype(int)
df["Semana"] = ((df["NumDia"] - 1) // 7) + 1
print("\nDías con mayor consumo y su semana correspondiente:")
print(df)
```

```python
plt.figure(figsize=(12, 6))
bars = plt.bar(df["Día"], df["Horas"], color="royalblue")
colors = {1: "red", 2: "green", 3: "orange", 4: "purple", 5: "blue"}
for i, semana in enumerate(df["Semana"]):
    bars[i].set_color(colors[semana])
    plt.title("Días con mayor consumo de redes sociales (superiores a la media)")
plt.xlabel("Días")
plt.ylabel("Horas de uso")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
