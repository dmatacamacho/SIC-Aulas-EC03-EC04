# Ejercicio: Análisis de horas de uso de redes sociales

Un grupo de estudiantes registró durante 30 días la **cantidad de horas que pasaron usando redes sociales**.
Los datos están representados en una **Serie de Pandas**, donde cada índice corresponde al día (`Día 1`, `Día 2`, ..., `Día 30`).

Su tarea será **analizar esta Serie** aplicando operaciones estadísticas y visualizaciones para entender mejor los hábitos de uso.

---

## Instrucciones

1. **Crear la Serie**

   * Genera una lista de 30 valores (pueden ser aleatorios o inventados entre 0 y 10 horas).
   * Convierte la lista en una Serie de Pandas.
   * Usa como índice `Día 1`, `Día 2`, ..., `Día 30`.

2. **Exploración de la Serie**

   * Muestra los primeros y últimos 5 valores.
   * Consulta el valor de un día específico (por ejemplo, `"Día 7"`).
   * Verifica la longitud, el tipo de datos y si existen valores faltantes.

3. **Estadísticas básicas**

   * Calcula: promedio, mediana, mínimo, máximo, varianza y desviación estándar.
   * Identifica los días con mayor y menor uso de redes sociales.

4. **Filtrado y condiciones**

   * Filtra los días donde se usaron **más de 6 horas**.
   * Cuenta cuántos días el uso fue **menor a 2 horas**.


5. **Visualizaciones**

   * Realiza al menos **2 gráficos diferentes** para representar la información.
   * Puedes elegir entre:

     * Gráfico de línea para ver la evolución diaria.
     * Histograma para la distribución de horas.
     * Gráfico de barras para comparar ciertos días.
   * ¡Se creativo! Elige los gráficos que mejor expliquen los resultados.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#crear la serie
plt.style.use('default')

np.random.seed(40)
horas = np.round(np.random.uniform(0, 10, 30), 1)

#indices desde dia 1----dia 30
dias = [f"Día {i}" for i in range(1, 31)]

#comvertir a serie con pandas
serie_zap = pd.Series(horas, index=dias, name="Horas de redes")

Tabla_datos = pd.DataFrame(horas, index=dias)

## Punto dos: mostrar datos
primeros_datos = Tabla_datos.head(5)
print("Los primeros 5 datos son:")
print(primeros_datos)

ultimos_datos = Tabla_datos.tail(5)
print("Los últimos 5 datos son:")
print(ultimos_datos)

dia_especifico = "Día 7"
valores_dia = Tabla_datos.loc[dia_especifico]
print(f"Valores del {dia_especifico}:")
print(valores_dia)

print("Dimensiones:")
print(Tabla_datos.shape)

print("Tipos de datos")
print(Tabla_datos.dtypes)

print("Valores faltantes:")
print(Tabla_datos.isnull().any())

print("Número total de valores faltantes:")
print(Tabla_datos.isnull().sum().sum())

# 5.- estadística básica
promedio = Tabla_datos.mean(axis=0)
mediana = Tabla_datos.median(axis=0)
minimo = Tabla_datos.min(axis=0)
maximo = Tabla_datos.max(axis=0)
varianza = Tabla_datos.var(axis=0)
desviacion = Tabla_datos.std(axis=0)

estadisticas = pd.DataFrame({
    "Promedio": promedio,
    "Mediana": mediana,
    "Mínimo": minimo,
    "Máximo": maximo,
    "Varianza": varianza,
    "Desviación": desviacion
})

print("Estadísticas por estudiante:")
print(estadisticas)

dias_mayor = Tabla_datos.idxmax(axis=0)
dias_menor = Tabla_datos.idxmin(axis=0)

print("Máximo uso de redes sociales")
print(dias_mayor)

print("Mínimo uso de redes sociales")
print(dias_menor)
```

```python
#Filtrado y condiciones

#Filtra los días donde se usaron más de 6 horas.
usomas6horas = serie_zap [serie_zap >= 6]
print("Los dias con mas de 6 horas de uso son")
print(usomas6horas)
```

```python

#Cuenta cuántos días el uso fue menor a 2 horas.
usomenos2horas = (serie_zap <= 2).sum()
print(f"Cantidad de dias que el uso fue menor a 2 horas fue: {usomenos2horas}")
```

```python
#graficos
#gráfico de lineas
plt.figure(figsize=(14, 6))
plt.plot(serie_zap.index, serie_zap.values, marker='o', linewidth=2, color='#1f77b4', label='Horas')
plt.title("Uso de redes sociales durante 30 dias", fontsize=16, fontweight='bold')
plt.xlabel("Dia", fontsize=12)
plt.ylabel("Horas", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

#gráfico de barras
dias_clave = ['Día 1', 'Día 7', 'Día 15', 'Día 21', 'Día 30']
valores_clave = serie_zap.loc[dias_clave]
colores = ['#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
plt.figure(figsize=(10, 6))
bars = plt.bar(dias_clave, valores_clave, color=colores)
plt.title("Comparacion de dias clave", fontsize=16, fontweight='bold')
plt.ylabel("Horas")
plt.grid(axis='y', alpha=0.3)
#etiquetas
for bar, val in zip(bars, valores_clave):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1,
             f'{val:.1f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```
