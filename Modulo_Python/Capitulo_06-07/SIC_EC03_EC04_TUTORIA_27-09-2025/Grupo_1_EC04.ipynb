# Ejercicio: Análisis de horas de uso de redes sociales

Un grupo de estudiantes registró durante 30 días la **cantidad de horas que pasaron usando redes sociales**.
Los datos están representados en una **Serie de Pandas**, donde cada índice corresponde al día (`Día 1`, `Día 2`, ..., `Día 30`).

Su tarea será **analizar esta Serie** aplicando operaciones estadísticas y visualizaciones para entender mejor los hábitos de uso.

---

## Instrucciones

1. **Crear la Serie**

   * Genera una lista de 30 valores (pueden ser aleatorios o inventados entre 0 y 10 horas).
   * Convierte la lista en una Serie de Pandas.
   * Usa como índice `Día 1`, `Día 2`, ..., `Día 30`.

2. **Exploración de la Serie**

   * Muestra los primeros y últimos 5 valores.
   * Consulta el valor de un día específico (por ejemplo, `"Día 7"`).
   * Verifica la longitud, el tipo de datos y si existen valores faltantes.

3. **Estadísticas básicas**

   * Calcula: promedio, mediana, mínimo, máximo, varianza y desviación estándar.
   * Identifica los días con mayor y menor uso de redes sociales.

4. **Filtrado y condiciones**

   * Filtra los días donde se usaron **más de 6 horas**.
   * Cuenta cuántos días el uso fue **menor a 2 horas**.


5. **Visualizaciones**

   * Realiza al menos **2 gráficos diferentes** para representar la información.
   * Puedes elegir entre:

     * Gráfico de línea para ver la evolución diaria.
     * Histograma para la distribución de horas.
     * Gráfico de barras para comparar ciertos días.
   * ¡Se creativo! Elige los gráficos que mejor expliquen los resultados.


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def calculo_estadisticas_basicas(serie_dias):
    """
    Calcula estadísticas básicas sobre el uso de redes sociales
    en un rango de 30 días calculando sobre sus horas.
    """

    promedio_horas = serie_dias.mean()
    mediana_horas = serie_dias.median()
    moda_horas = serie_dias.mode()

    minimo_horas = serie_dias.min()
    maximo_horas = serie_dias.max()
    rango = maximo_horas - minimo_horas
    varianza = serie_dias.var()
    desviacion_estandar = serie_dias.std()

    print("="*50)
    print("ESTADISTICAS BASICAS")
    print("="*50)

    print(f"Promedio de uso: {promedio_horas:.2f} horas")
    print(f"Mediana de horas: {mediana_horas:.2f} horas")
    print(f"Moda: {moda_horas.iloc[0]} horas\n")

    print("="*50)
    print("CALCULO CON HORAS EN REDES SOCIALES")
    print("="*50)

    print(f"Mínimo: {minimo_horas} horas")
    print(f"Máximo: {maximo_horas} horas")
    print(f"Rango: {rango:.2f} horas")
    print(f"Varianza: {varianza:.2f}")
    print(f"Desviación estándar: {desviacion_estandar:.2f} horas\n")

    dia_maximo = serie_dias.idxmax()
    valor_maximo = serie_dias.max()

    dia_minimo = serie_dias.idxmin()
    valor_minimo = serie_dias.min()

    print("="*50)
    print("DIAS CON MAYOR Y MENOR USO DE REDES SOCIALES")
    print("="*50)

    print(f"Día con mayor uso: {dia_maximo} con {valor_maximo} horas")
    print(f"Día con menor uso: {dia_minimo} con {valor_minimo} horas")

    estadisticas_basicas = {
        'promedio': promedio_horas,
        'mediana': mediana_horas,
        'moda': moda_horas.iloc[0],
        'minimo': minimo_horas,
        'maximo': maximo_horas,
        'varianza': varianza,
        'desviacion_estandar': desviacion_estandar,
        'dia_maximo': dia_maximo,
        'dia_minimo': dia_minimo
    }

    return estadisticas_basicas

lista_dias = {}
for i in range(1, 31):
    lista_dias[f"Dia {i}"] = np.random.randint(1, 11)

serie_dias = pd.Series(lista_dias)

print("="*50)
print("SERIE DE HORAS DE USO DE REDES SOCIALES")
print("="*50)
print(serie_dias)

print("\nPrimeros 5 valores:")
print(serie_dias.head())

print("\nÚltimos 5 valores:")
print(serie_dias.tail())

print("\nValor de Dia 7:")
print(serie_dias["Dia 7"])
print("\n")
print("="*50)
print("DATOS DE LA SERIE")
print("="*50)

print("\nLongitud de la serie:", len(serie_dias))

print("Tipo de datos de la serie" , serie_dias.dtype)

print("¿Existen valores faltantes ,?", serie_dias.isnull().any)

estadisticas_resultantes = calculo_estadisticas_basicas(serie_dias)
print(estadisticas_resultantes)

dias_mas_6h = serie_dias[serie_dias > 6]
print("\nDías con más de 6 horas:")
print(dias_mas_6h)

dias_menos_2h = (serie_dias < 2).sum()
print(f"\nDías con menos de 2 horas: {dias_menos_2h}")

plt.style.use('fivethirtyeight')

fig, (ax1, ax2) = plt.subplots(2, figsize = (15,15))

ax1.hist(serie_dias, color = 'orange')

ax1.set_title('Histograma días y horas de uso redes sociales')
ax1.tick_params(axis = 'x', rotation = 45)

colors = ['red' if value > serie_dias.mean() else 'purple' for value in serie_dias]
ax2.bar(serie_dias.index, serie_dias.values, color = colors)
ax2.axhline(y=serie_dias.mean(), color='r', linestyle='--', label=f'Promedio: {serie_dias.mean():.2f}')

ax2.set_title('Diagrama de barras de uso redes sociales')
ax2.tick_params(axis = 'x', rotation = 45)

plt.tight_layout()
plt.show()
```
