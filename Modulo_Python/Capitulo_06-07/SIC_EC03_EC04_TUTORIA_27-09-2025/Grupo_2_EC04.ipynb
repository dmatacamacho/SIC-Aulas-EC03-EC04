# Ejercicio: Análisis de horas de uso de redes sociales

Un grupo de estudiantes registró durante 30 días la **cantidad de horas que pasaron usando redes sociales**.
Los datos están representados en una **Serie de Pandas**, donde cada índice corresponde al día (`Día 1`, `Día 2`, ..., `Día 30`).

Su tarea será **analizar esta Serie** aplicando operaciones estadísticas y visualizaciones para entender mejor los hábitos de uso.

---

## Instrucciones

1. **Crear la Serie**

   * Genera una lista de 30 valores (pueden ser aleatorios o inventados entre 0 y 10 horas).
   * Convierte la lista en una Serie de Pandas.
   * Usa como índice `Día 1`, `Día 2`, ..., `Día 30`.

2. **Exploración de la Serie**

   * Muestra los primeros y últimos 5 valores.
   * Consulta el valor de un día específico (por ejemplo, `"Día 7"`).
   * Verifica la longitud, el tipo de datos y si existen valores faltantes.

3. **Estadísticas básicas**

   * Calcula: promedio, mediana, mínimo, máximo, varianza y desviación estándar.
   * Identifica los días con mayor y menor uso de redes sociales.

4. **Filtrado y condiciones**

   * Filtra los días donde se usaron **más de 6 horas**.
   * Cuenta cuántos días el uso fue **menor a 2 horas**.


5. **Visualizaciones**

   * Realiza al menos **2 gráficos diferentes** para representar la información.
   * Puedes elegir entre:

     * Gráfico de línea para ver la evolución diaria.
     * Histograma para la distribución de horas.
     * Gráfico de barras para comparar ciertos días.
   * ¡Se creativo! Elige los gráficos que mejor expliquen los resultados.


```python
import pandas as pd
import numpy as np
import random as rd
import matplotlib.pyplot as plt
```

```python
''' Crear la Serie '''
def crear_serie():
  horas= np.random.uniform(0,10,30).round(2)
  indices = [f"Dia {i}" for i in range (1,31)]
  return pd.Series(horas,index=indices)
```

```python
'''Exploración de la serie'''

def explorar_serie(serie , dia_especifico):
    print("\nExploración de la serie: ")
    print('-'*30)
    print("Primeros 5 valores:")
    print(serie.head())
    print("\nÚltimos 5 valores:")
    print(serie.tail())
    if dia_especifico in serie.index:
        print(f"\nValor del {dia_especifico}: {serie.loc[dia_especifico]}")
    else:
        print(f"\nEl día {dia_especifico} no está en la serie.")
    print("\nLongitud de la serie:", len(serie))
    print("Tipo de datos:", serie.dtype)
    print("¿Existen valores faltantes?", "Si" if serie.isnull().any()==True else "No")
    print("Faltan ", serie.isnull().sum(), "valores.\n")
```

```python
''' Estadisticas Básicas '''

def estadisticas_basicas(serie):
  print(f"Estadísticas obtenidas de la serie: ")
  print('-'*30)
  print(f"Promedio: {serie.mean():.2f}")
  print(f"Mediana: {serie.median()}")
  print(f"Mínimo: {serie.min()}")
  print(f"Máximo: {serie.max()}")
  print(f"Varianza: {serie.var():.2f}")
  print(f"Desviación estándar: {serie.std():.2f}")

  print(f"Los dias con mayor uso de las Redes Sociales fueron: {serie.idxmax()} con un total de {serie.max()} horas")
  print(f"Los dias con menor uso de las Redes Sociales fueron: {serie.idxmin()} con un total de {serie.min()} horas")
```

```python
def filtro(serie):
  """Función que filtra los días con menos de 6 Horas de uso y cuentas los días con uso de RS menor a 2 horas"""
  filtro= serie>6
  horas_mayores_6= serie[filtro]
  print("\n--- Días con mas de 6 Horas de uso de Redes Sociales ---")
  print(horas_mayores_6)
  dias_uso_menores2Horas= (serie < 2).sum()
  print("\n--- 2. Conteo de Días ---")
  print(f"Total de días donde el uso fue menor a 2 horas: {dias_uso_menores2Horas} días")
```

```python
def graficos_estadisticos(serie):
  plt.style.use('seaborn-v0_8')
  fig, ((ax1,ax2,),(ax3,ax4)) = plt.subplots(2,2,figsize=(15,10))
  fig.suptitle('Analisis del uso de redes sociales')

  serie.plot(kind='bar', ax=ax1, color='orange')
  ax1.set_title('Comparación de días')
  ax1.set_xlabel('Días')
  ax1.set_ylabel('Horas')
  ax1.tick_params(axis='x', rotation=45)

  serie.plot(kind='hist', bins=10, ax=ax2, edgecolor='black')
  ax2.set_title('Distribución de horas')
  ax2.set_xlabel('Horas')
  ax2.set_ylabel('Frecuencia')

  ax3.boxplot(serie, vert=False, patch_artist=True, boxprops=dict(facecolor='skyblue'))
  ax3.set_title('Boxplot de horas')
  ax3.set_xlabel('Horas')

  ax4.plot(serie, marker='o', linestyle='-', color='g')
  ax4.set_title('Evolucion diaria')
  ax4.set_xlabel('Días')
  ax4.set_ylabel('Horas')
  ax4.tick_params(axis='x', rotation=45)

  plt.tight_layout()
  plt.show()
```

```python
serie = crear_serie()
print("Registro de dias en base a la cantidad de horas que pasaron usando redes sociales:")
print("-"*83)
print(serie)
print("\n")
explorar_serie(serie, "Dia 7")

estadisticas_basicas(serie)
filtro(serie)
graficos_estadisticos(serie)
```
