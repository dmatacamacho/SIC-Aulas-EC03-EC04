# Ejercicio: Análisis de horas de uso de redes sociales

Un grupo de estudiantes registró durante 30 días la **cantidad de horas que pasaron usando redes sociales**.
Los datos están representados en una **Serie de Pandas**, donde cada índice corresponde al día (`Día 1`, `Día 2`, ..., `Día 30`).

Su tarea será **analizar esta Serie** aplicando operaciones estadísticas y visualizaciones para entender mejor los hábitos de uso.

---

## Instrucciones

1. **Crear la Serie**

   * Genera una lista de 30 valores (pueden ser aleatorios o inventados entre 0 y 10 horas).
   * Convierte la lista en una Serie de Pandas.
   * Usa como índice `Día 1`, `Día 2`, ..., `Día 30`.

2. **Exploración de la Serie**

   * Muestra los primeros y últimos 5 valores.
   * Consulta el valor de un día específico (por ejemplo, `"Día 7"`).
   * Verifica la longitud, el tipo de datos y si existen valores faltantes.

3. **Estadísticas básicas**

   * Calcula: promedio, mediana, mínimo, máximo, varianza y desviación estándar.
   * Identifica los días con mayor y menor uso de redes sociales.

4. **Filtrado y condiciones**

   * Filtra los días donde se usaron **más de 6 horas**.
   * Cuenta cuántos días el uso fue **menor a 2 horas**.


5. **Visualizaciones**

   * Realiza al menos **2 gráficos diferentes** para representar la información.
   * Puedes elegir entre:

     * Gráfico de línea para ver la evolución diaria.
     * Histograma para la distribución de horas.
     * Gráfico de barras para comparar ciertos días.
   * ¡Se creativo! Elige los gráficos que mejor expliquen los resultados.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#1 Crear la Serie
horas= np.random.randint(0,11, size=30) #30 valores entre 0 y 10
dias=[]
for i in range(1,31):
  dias.append(f"Dia {i}")

serieRedes=pd.Series(horas, index=dias)

print("Datos registrados:")
print(serieRedes)

#2 Exploración de la Serie
print("\nPrimeros 5 valores:")
print(serieRedes.head())
print("\nUltimos 5 valores:")
print(serieRedes.tail())

print("\nValor dia 7:")
print(serieRedes["Dia 7"])

print("Longitud:",len(serieRedes))
print("Tipo de datos:",serieRedes.dtype)
print("Faltan valores?",serieRedes.isna().any())

#3 Estadísticas básicas
print("Promedio:", serieRedes.mean())
print("Mediana:", serieRedes.median())
print("Minimo:", serieRedes.min())
print("Maximo:", serieRedes.max())
print("Varianza:", serieRedes.var())
print("Desviación estandar:", serieRedes.std())

diaMayor = serieRedes.idxmax()
diaMenor = serieRedes.idxmin()

print("\nDia con mayor uso:", diaMayor, ":", serieRedes[diaMayor])
print("Dia con menor uso:", diaMenor, ":", serieRedes[diaMenor])

#4
print("Dias con mas de 6 horas de uso:")
print(serieRedes[serieRedes > 6])

diasMenor2 = (serieRedes < 2).sum()
print("\nNúmero de días con menos de 2 horas de uso:", diasMenor2)

```

# Visualizaciones

#Gráfico de línea

```python
#Gráfico de línea para ver la evolución diaria.
plt.figure(figsize=(10, 4))
plt.plot(
    serieRedes.index,
    serieRedes.values,
    marker='o',
    linewidth=2,
    color='#2a9d8f',
)
plt.title('Evolución diaria', fontsize=12)
plt.xlabel('Días')
plt.ylabel('Horas')
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(rotation=45, ha='right')
plt.ylim(0, max(serieRedes.values) + 1)

plt.tight_layout()
plt.show()

print()
```

#Gráfico de barras.

```python
#Gráfico de barras.
dias_comparar = ['Dia 1', 'Dia 7', 'Dia 15', 'Dia 22', 'Dia 30']

dias_seleccion = [d for d in dias_comparar if d in serieRedes.index]
valores = serieRedes.loc[dias_seleccion]

plt.figure(figsize=(8, 4))
colores = ['#457b9d', '#e76f51', '#2a9d8f', '#f4a261', '#1d3557']
plt.bar(dias_seleccion, valores, color=colores[:len(dias_seleccion)])
plt.title('Uso de redes: comparación de días', fontsize=12)
plt.xlabel('Días')
plt.ylabel('Horas')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.xticks(rotation=0)

for x, y in zip(dias_seleccion, valores):
    plt.text(x, y + 0.1, str(int(y)), ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
