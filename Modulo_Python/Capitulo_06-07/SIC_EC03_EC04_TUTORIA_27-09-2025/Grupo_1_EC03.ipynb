# Ejercicio: Análisis de horas de uso de redes sociales

Un grupo de estudiantes registró durante 30 días la **cantidad de horas que pasaron usando redes sociales**.
Los datos están representados en una **Serie de Pandas**, donde cada índice corresponde al día (`Día 1`, `Día 2`, ..., `Día 30`).

Su tarea será **analizar esta Serie** aplicando operaciones estadísticas y visualizaciones para entender mejor los hábitos de uso.

---

## Instrucciones

1. **Crear la Serie**

   * Genera una lista de 30 valores (pueden ser aleatorios o inventados entre 0 y 10 horas).
   * Convierte la lista en una Serie de Pandas.
   * Usa como índice `Día 1`, `Día 2`, ..., `Día 30`.

2. **Exploración de la Serie**

   * Muestra los primeros y últimos 5 valores.
   * Consulta el valor de un día específico (por ejemplo, `"Día 7"`).
   * Verifica la longitud, el tipo de datos y si existen valores faltantes.

3. **Estadísticas básicas**

   * Calcula: promedio, mediana, mínimo, máximo, varianza y desviación estándar.
   * Identifica los días con mayor y menor uso de redes sociales.

4. **Filtrado y condiciones**

   * Filtra los días donde se usaron **más de 6 horas**.
   * Cuenta cuántos días el uso fue **menor a 2 horas**.


5. **Visualizaciones**

   * Realiza al menos **2 gráficos diferentes** para representar la información.
   * Puedes elegir entre:

     * Gráfico de línea para ver la evolución diaria.
     * Histograma para la distribución de horas.
     * Gráfico de barras para comparar ciertos días.
   * ¡Se creativo! Elige los gráficos que mejor expliquen los resultados.


```python
# Resolucion
import random, pandas as pd, matplotlib.pyplot as plt, numpy as np
import numpy as np
from sklearn.linear_model import LinearRegression

print("\n" + "="*50)
print("Crear la Serie:")
print("-"*50 + "\n")
# Creacion de listas acorden a la generacion de datos
horas_redes_sociales = [random.randint(0, 10) for _ in range(30)]
horas_estudio = [random.randint(0, 6) for _ in range(30)]
dias = [f"Día {i+1}" for i in range(30)]

#Creacion de la serie en pandas conforme a las listas anteriores
serie_horas_redes_sociales = pd.Series(horas_redes_sociales, index=dias)
serie_horas_estudio = pd.Series(horas_estudio, index=dias)
#Exploracion de la serie
print("Primeros 5 valores")
print(serie_horas_redes_sociales[:5])
print("Ultimos 5 valores")
print(serie_horas_redes_sociales[-5:])
print(f"Dia 28: {serie_horas_redes_sociales['Día 28']}")

print(f"Longitud de la serie: {len(serie_horas_redes_sociales)}")
print(f"Tipo de datos de los valores: {serie_horas_redes_sociales.dtype}")
valores_faltantes_redes = serie_horas_redes_sociales.isnull().sum()
print(f"Valores faltantes: {valores_faltantes_redes}")


```

```python
print("\n" + "="*50)
print("Estadisticas del Uso de Redes Sociales:")
print("-"*50 + "\n")

promedio = f"Promedio {serie_horas_redes_sociales.mean():.2f} horas"
mediana = f"Mediana {serie_horas_redes_sociales.median():.2f} horas"
minimo = f"Minimo {serie_horas_redes_sociales.min():.2f}"
maximo = f"Maximo {serie_horas_redes_sociales.max():.2f}"
varianza = f"Varianza {serie_horas_redes_sociales.var():.2f}"
desviacion_estandar= f"Desviacion estandar {serie_horas_redes_sociales.std():.2f} horas"

estadisticas = [promedio, mediana, minimo, maximo, varianza, desviacion_estandar]
for estadistica in estadisticas:
  print(estadistica)

print("Dias con mayor numero de horas")
max_horas = serie_horas_redes_sociales.max()
dias_max = serie_horas_redes_sociales[serie_horas_redes_sociales == max_horas]
for dia, horas in dias_max.items():
    print(f"• {dia}: {horas} horas")

print("Dias con menor numero de horas")
min_horas = serie_horas_redes_sociales.min()
dias_min = serie_horas_redes_sociales[serie_horas_redes_sociales == min_horas]
for dia, horas in dias_min.items():
    print(f"• {dia}: {horas} horas")

```

```python
print("\n" + "="*50)
print("Filtrado y condiciones")
print("-"*50 + "\n")
print("Dias de mas de 6 horas")
for index, valor in serie_horas.items():
    if valor> 6:
      print(index)
print("Cantidad de Dias de menos de 2 horas")
contador=0
for index, valor in serie_horas.items():
    if valor < 2:
      contador=contador +1
print(f"Son { contador} dias")
```

```python
# Matplotlib
print("\n" + "="*50)
print("VISUALIZACIONES")
print("-"*50 + "\n")

plt.style.use('seaborn-v0_8')


fig = plt.figure(figsize=(15, 12))


gs = plt.GridSpec(2, 2, height_ratios=[1, 1.2], figure=fig)


ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, :])

fig.suptitle("Análisis de horas de uso de redes sociales vs horas de estudio", fontsize=16)
promedio_redes = serie_horas_redes_sociales.mean()
serie_horas_redes_sociales.plot(ax=ax1, kind='line', marker='o', color='blue',
                               linewidth=2, markersize=6, label='Horas Redes Sociales')
ax1.axhline(y=promedio_redes, color='red', linestyle='--', linewidth=2,
           label=f'Promedio ({promedio_redes:.2f} horas)')
for i, (dia, valor) in enumerate(serie_horas_redes_sociales.items()):
    if i % 5 == 0:  # Mostrar etiquetas cada 5 días
        ax1.annotate(dia, (i, valor), xytext=(8, 8), textcoords='offset points',
                    fontsize=8, bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))

ax1.set_title("Horas de Redes Sociales con Línea de Promedio", fontsize=12, fontweight="bold")
ax1.set_xlabel('Días')
ax1.set_ylabel('Horas')
ax1.tick_params(axis='x', rotation=45)
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_facecolor('#f8f9fa')



X = np.array(range(len(serie_horas_redes_sociales))).reshape(-1, 1)
y = serie_horas_redes_sociales.values

modelo = LinearRegression()
modelo.fit(X, y)
y_pred = modelo.predict(X)


ax2.scatter(X, y, color='blue', alpha=0.7, s=60, label='Datos reales')
ax2.plot(X, y_pred, color='red', linewidth=3, label='Línea de regresión')


ecuacion = f'y = {modelo.coef_[0]:.2f}x + {modelo.intercept_:.2f}'
r_cuadrado = f'R² = {modelo.score(X, y):.3f}'

ax2.text(0.05, 0.95, ecuacion, transform=ax2.transAxes, fontsize=11,
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.9))
ax2.text(0.05, 0.85, r_cuadrado, transform=ax2.transAxes, fontsize=11,
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.9))

ax2.set_title("Regresión Lineal: Horas en Redes Sociales", fontsize=12, fontweight="bold")
ax2.set_xlabel('Días (secuencia)')
ax2.set_ylabel('Horas')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_facecolor('#f8f9fa')


df_comparativo = pd.DataFrame({
    'Redes Sociales': serie_horas_redes_sociales,
    'Estudio': serie_horas_estudio
})


ancho = 0.35
x = range(len(df_comparativo))

ax3.bar([i - ancho/2 for i in x], df_comparativo['Redes Sociales'],
        width=ancho, label='Redes Sociales', color='skyblue', edgecolor='black', alpha=0.8)
ax3.bar([i + ancho/2 for i in x], df_comparativo['Estudio'],
        width=ancho, label='Estudio', color='lightcoral', edgecolor='black', alpha=0.8)

ax3.set_title("Comparación Completa: Horas de Redes Sociales vs Horas de Estudio (30 días)",
              fontsize=14, fontweight="bold")
ax3.set_xlabel('Días')
ax3.set_ylabel('Horas')
ax3.set_xticks(x)
ax3.set_xticklabels(dias, rotation=45, fontsize=9)
ax3.legend()
ax3.grid(True, alpha=0.3)
ax3.set_facecolor('#f8f9fa')


for i, (red, est) in enumerate(zip(df_comparativo['Redes Sociales'], df_comparativo['Estudio'])):
    if red > 6:
        ax3.text(i - ancho/2, red + 0.1, f'{red}h', ha='center', va='bottom', fontsize=8, fontweight='bold')
    if est > 3:
        ax3.text(i + ancho/2, est + 0.1, f'{est}h', ha='center', va='bottom', fontsize=8, fontweight='bold')

plt.tight_layout()
plt.subplots_adjust(top=0.94)
plt.show()
```
