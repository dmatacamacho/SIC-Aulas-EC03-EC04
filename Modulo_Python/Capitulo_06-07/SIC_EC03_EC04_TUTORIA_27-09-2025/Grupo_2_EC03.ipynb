# Ejercicio: Análisis de horas de uso de redes sociales

Un grupo de estudiantes registró durante 30 días la **cantidad de horas que pasaron usando redes sociales**.
Los datos están representados en una **Serie de Pandas**, donde cada índice corresponde al día (`Día 1`, `Día 2`, ..., `Día 30`).

Su tarea será **analizar esta Serie** aplicando operaciones estadísticas y visualizaciones para entender mejor los hábitos de uso.

---

## Instrucciones

1. **Crear la Serie**

   * Genera una lista de 30 valores (pueden ser aleatorios o inventados entre 0 y 10 horas).
   * Convierte la lista en una Serie de Pandas.
   * Usa como índice `Día 1`, `Día 2`, ..., `Día 30`.

2. **Exploración de la Serie**

   * Muestra los primeros y últimos 5 valores.
   * Consulta el valor de un día específico (por ejemplo, `"Día 7"`).
   * Verifica la longitud, el tipo de datos y si existen valores faltantes.

3. **Estadísticas básicas**

   * Calcula: promedio, mediana, mínimo, máximo, varianza y desviación estándar.
   * Identifica los días con mayor y menor uso de redes sociales.

4. **Filtrado y condiciones**

   * Filtra los días donde se usaron **más de 6 horas**.
   * Cuenta cuántos días el uso fue **menor a 2 horas**.


5. **Visualizaciones**

   * Realiza al menos **2 gráficos diferentes** para representar la información.
   * Puedes elegir entre:

     * Gráfico de línea para ver la evolución diaria.
     * Histograma para la distribución de horas.
     * Gráfico de barras para comparar ciertos días.
   * ¡Se creativo! Elige los gráficos que mejor expliquen los resultados.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Crear la Serie
np.random.seed(42)
horas_uso = np.random.randint(0, 11, 30)
indices_dias = [f'Día {i}' for i in range(1, 31)]
serie_horas = pd.Series(horas_uso, index=indices_dias)
print("Creación de la Serie")
print("Serie de horas de uso de redes sociales:")
print(serie_horas)
print("\n" + "="*50 + "\n")


# 2. Exploración de la Serie
print("### 2. Exploración de la Serie ###")
print("Primeros 5 días:")
print(serie_horas.head())
print("-" * 20)
print("Últimos 5 días:")
print(serie_horas.tail())
print("-" * 20)

dia_especifico = "Día 7"
print(f"Horas de uso el {dia_especifico}: {serie_horas[dia_especifico]} horas")
print("-" * 20)

print(f"Longitud de la Serie: {len(serie_horas)} días")
print(f"Tipo de datos: {serie_horas.dtype}")
print(f"Valores faltantes: {serie_horas.isnull().sum()}")
print("\n" + "="*50 + "\n")


# 3. Estadísticas básicas
print("### 3. Estadísticas Básicas ###")
print(f"Promedio de horas: {serie_horas.mean():.2f}")
print(f"Mediana de horas: {serie_horas.median()}")
print(f"Mínimo de horas: {serie_horas.min()}")
print(f"Máximo de horas: {serie_horas.max()}")
print(f"Varianza: {serie_horas.var():.2f}")
print(f"Desviación estándar: {serie_horas.std():.2f}")
print("-" * 20)

print(f"Día con mayor uso: {serie_horas.idxmax()} ({serie_horas.max()} horas)")
print(f"Día con menor uso: {serie_horas.idxmin()} ({serie_horas.min()} horas)")
print("\n" + "="*50 + "\n")


# 4. Filtrado y condiciones
print("### 4. Filtrado y Condiciones ###")
dias_mas_de_6_horas = serie_horas[serie_horas > 6]
print("Días con más de 6 horas de uso:")
print(dias_mas_de_6_horas)
print("-" * 20)

dias_menos_de_2_horas = (serie_horas < 2).sum()
print(f"Número de días con menos de 2 horas de uso: {dias_menos_de_2_horas}")
print("\n" + "="*50 + "\n")


# 5. Visualizaciones
print("### 5. Visualizaciones ###")
print("Generando gráficos...")

plt.figure(figsize=(12, 6)) # Define el tamaño de la figura
serie_horas.plot(kind='line', marker='o', linestyle='-', color='b')
plt.style.use('ggplot')
plt.title('Evolución del Uso de Redes Sociales Durante 30 Días')
plt.xlabel('Día')
plt.ylabel('Horas de Uso')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Gráfico 2: Histograma para la distribución de horas ---
plt.figure(figsize=(10, 6))
bins = range(serie_horas.max() + 2)
serie_horas.plot(kind='hist', bins=bins, color='skyblue', edgecolor='black')
plt.style.use('ggplot')
plt.title('Distribución de las Horas de Uso de Redes Sociales')
plt.xlabel('Rango de Horas')
plt.ylabel('Frecuencia (Número de Días)')
plt.xticks(bins)
plt.grid(axis='y', alpha=0.75)
plt.show()

print("Análisis completado.")
```

```python
import seaborn as sns


valores = list(serie_horas.values) + [np.nan] * (35 - len(serie_horas))
matriz_calendario = np.array(valores).reshape(5, 7)
plt.figure(figsize=(10, 6))
sns.heatmap(
    matriz_calendario,
    annot=True, fmt=".0f", cmap="RdYlGn_r", linewidths=0.5,
    cbar_kws={'label': 'Horas de uso'}
)
plt.title("Heatmap estilo calendario - Uso de redes sociales (30 días)")
plt.xlabel("Días de la semana")
plt.ylabel("Semanas")
plt.yticks(rotation=0)
plt.xticks(
    ticks=np.arange(7)+0.5,
    labels=["Lun","Mar","Mié","Jue","Vie","Sáb","Dom"],
    rotation=0
)
plt.show()
```

```python

#gráfico torta
categoria = {
    "USO LIGERO (MENOR 2H)": (serie_horas < 2).sum(),
    "USO MODERADO (2H-6H)": ((serie_horas >= 2) & (serie_horas <= 6)).sum(),
    "USO ALTO (MAYOR 6H)": (serie_horas > 9).sum()
}

plt.figure(figsize=(8, 6))
plt.pie(
    categoria.values(),
    labels=categoria.keys(),
    autopct='%1.1f%%',
    startangle=90,
    colors=['#8fd9a8', '#ffd966', '#ff6f69'],
    wedgeprops=dict(width=0.4)
)
plt.title('Distribución de Categorías de Uso de Redes Sociales')
plt.show()
```
