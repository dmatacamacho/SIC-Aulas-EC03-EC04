```python
"""
Dado un laberinto representado como una matriz (0 = camino, 1 = pared), 
implementa un algoritmo que encuentre una ruta desde la esquina superior 
izquierda hasta la esquina inferior derecha, moviéndose solo en cuatro 
direcciones (arriba, abajo, izquierda, derecha).
"""

def resolver_laberinto(laberinto, x, y, solucion):
    n = len(laberinto)

    # Caso base: llegamos al destino
    if x == n-1 and y == n-1:
        solucion[x][y] = 1
        return True

    # Validar movimiento
    if x >= 0 and y >= 0 and x < n and y < n and laberinto[x][y] == 0:
        if solucion[x][y] == 1:
            return False

        # Marcar paso actual
        solucion[x][y] = 1

        # Explorar en 4 direcciones
        if resolver_laberinto(laberinto, x+1, y, solucion):
            return True
        if resolver_laberinto(laberinto, x, y+1, solucion):
            return True
        if resolver_laberinto(laberinto, x-1, y, solucion):
            return True
        if resolver_laberinto(laberinto, x, y-1, solucion):
            return True

        # Backtracking
        solucion[x][y] = 0
        return False

    return False

# Ejemplo de ejecución
laberinto = [
    [0, 1, 0, 0],
    [0, 0, 0, 1],
    [1, 0, 1, 0],
    [0, 0, 0, 0]
]
n = len(laberinto)
solucion = [[0]*n for _ in range(n)]

if resolver_laberinto(laberinto, 0, 0, solucion):
    for fila in solucion:
        print(fila)
else:
    print("No hay solución")

```

```python

"""
Ejercicio de N- reinas

"""
def es_valido(tablero, fila, col, n):
    # Verificar columna
    for i in range(fila):
        if tablero[i][col] == 1:
            return False

    # Diagonal superior izquierda
    for i, j in zip(range(fila, -1, -1), range(col, -1, -1)):
        if tablero[i][j] == 1:
            return False

    # Diagonal superior derecha
    for i, j in zip(range(fila, -1, -1), range(col, n)):
        if tablero[i][j] == 1:
            return False

    return True

def resolver_n_reinas(tablero, fila, n):
    if fila >= n:
        return True

    for col in range(n):
        if es_valido(tablero, fila, col, n):
            tablero[fila][col] = 1
            
            if resolver_n_reinas(tablero, fila+1, n):
                return True
            
            tablero[fila][col] = 0

    return False

# Inicialización y ejecución
n = 4
tablero = [[0]*n for _ in range(n)]

if resolver_n_reinas(tablero, 0, n):
    for fila in tablero:
        print(fila)
else:
    print("No hay solución")
```
